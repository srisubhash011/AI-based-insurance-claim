```python
import cv2
import os
import numpy as np
import random

real_train = r"C:\Users\HP\Desktop\desktop\OPEN LAB\OPENLAB_PROJECT\data\ai_dataset\train\real"
ai_train = r"C:\Users\HP\Desktop\desktop\OPEN LAB\OPENLAB_PROJECT\data\ai_dataset\train\ai"

real_val = r"C:\Users\HP\Desktop\desktop\OPEN LAB\OPENLAB_PROJECT\data\ai_dataset\val\real"
ai_val = r"C:\Users\HP\Desktop\desktop\OPEN LAB\OPENLAB_PROJECT\data\ai_dataset\val\ai"

os.makedirs(ai_train, exist_ok=True)
os.makedirs(ai_val, exist_ok=True)


def manipulate(img):

    choice = random.randint(0,4)

    if choice == 0:
        img = cv2.GaussianBlur(img,(11,11),0)

    elif choice == 1:
        img = cv2.convertScaleAbs(img,alpha=1.4,beta=40)

    elif choice == 2:
        noise = np.random.normal(0,25,img.shape).astype(np.uint8)
        img = cv2.add(img,noise)

    elif choice == 3:
        img = cv2.flip(img,1)

    elif choice == 4:
        img = cv2.resize(img,(256,256))
        img = cv2.resize(img,(224,224))

    return img


def create_ai_images(real_folder, ai_folder):

    for img_name in os.listdir(real_folder):

        path = os.path.join(real_folder,img_name)

        img = cv2.imread(path)

        if img is None:
            continue

        fake = manipulate(img)

        save_path = os.path.join(ai_folder,img_name)

        cv2.imwrite(save_path,fake)


print("Generating AI train images...")
create_ai_images(real_train,ai_train)

print("Generating AI validation images...")
create_ai_images(real_val,ai_val)

print("Done")
```
